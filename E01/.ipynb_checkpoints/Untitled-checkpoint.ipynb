```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x, C, a):
    return x.T @ C @ x

def f2(x, C, a):
    return np.exp(-0.5 * f1(x, C, a))

def f3(x, C, a):
    return f1(x, C, a) / (a**2 + f1(x, C, a))

def c_diag(n, c=10):
    return np.diag([pow(c, (i-1)/(n-1)) for i in range(n)])
```

```python
def plot_func(func):
    x = np.linspace(-1, 1, 10)
    X, Y = np.meshgrid(x, x)
    a = 0.1
    res = list()
    for i in x:
        for j in x:
            res.append(func(np.array([i, j]), c_diag(2), a))
    res = np.array(res).reshape(X.shape)
    
    fig = plt.figure(figsize=(15, 5))
    ax1 = fig.add_subplot(131, projection='3d')
    ax1.plot_surface(X, Y, res, cmap='viridis')
```

```python
plot_func(f1)
plot_func(f2)
plot_func(f3)
```
